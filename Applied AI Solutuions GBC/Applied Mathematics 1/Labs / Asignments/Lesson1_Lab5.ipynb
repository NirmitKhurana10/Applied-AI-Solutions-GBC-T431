## Lab Excercise 5

The process below shows how to implement a simple linear regression analysis for a vehicle pricing scenario. Re-run it and take note of the R-sqaured value. Then. change the std value in np.random.normal to 1, 5 and 10. Observe the change in R-sqaured. How do you explain the change in R-sqaured value?

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

vehicle_age = np.random.uniform(0,10,1000)
vehicle_price = 70000 - (vehicle_age + np.random.normal(0, 1, 1000)) * 3000

vehicle_age = vehicle_age.reshape(-1, 1)  #reshape for single feature, use (1,-1) for single sample
vehicle_price = vehicle_price.reshape(-1, 1)

plt.scatter(vehicle_age, vehicle_price)
```

```python
reg = LinearRegression().fit(vehicle_age, vehicle_price)
reg.score(vehicle_age, vehicle_price)
```

## STD : 5

```python
vehicle_age = np.random.uniform(0,10,1000)
vehicle_price_5 = 70000 - (vehicle_age + np.random.normal(0, 5, 1000)) * 3000

vehicle_age = vehicle_age.reshape(-1, 1)  #reshape for single feature, use (1,-1) for single sample
vehicle_price_5 = vehicle_price_5.reshape(-1, 1)

plt.scatter(vehicle_age, vehicle_price_5)

reg = LinearRegression().fit(vehicle_age, vehicle_price_5)
print("New R squared Value with std 5: ", reg.score(vehicle_age, vehicle_price_5))
```

## STD : 10

```python
vehicle_age = np.random.uniform(0,10,1000)
vehicle_price_10 = 70000 - (vehicle_age + np.random.normal(0, 10, 1000)) * 3000

vehicle_age = vehicle_age.reshape(-1, 1)  #reshape for single feature, use (1,-1) for single sample
vehicle_price_10 = vehicle_price_10.reshape(-1, 1)

plt.scatter(vehicle_age, vehicle_price_10)

reg = LinearRegression().fit(vehicle_age, vehicle_price_10)
print("New R squared Value with std 5: ", reg.score(vehicle_age, vehicle_price_10))
```

##### As seen in the question, we changed the Standard Deviation for the data. A larger standard deviation means more noise. The data points are more spread out and less consistent with the true relationship.


#### Original Standard Deviation (std = 1)
### R-squared value: 0.891
### -----------------------
##### With a small standard deviation of 0.001, there is very little noise added to  data. As a result:
##### The data points are close to the true underlying relationship.
##### The linear regression model can fit the data well, explaining a high proportion of the variance in vehicle prices.

# --------------------------------------


#### Increased Standard Deviation (std = 5)
### R-squared value: 0.239
### -----------------------
##### When I increase the standard deviation to 5:
##### The noise in the data increases significantly.
##### Data points become more scattered and less consistent with the true relationship.
##### The linear regression model struggles to fit this noisy data, and its ability to explain the variance decreases, leading to a lower R-squared value.
