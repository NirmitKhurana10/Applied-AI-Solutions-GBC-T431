```python
#### Generating data set and showing how will it look like


%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

vehicle_age = np.random.uniform(0,10,1000)
vehicle_price = 70000 - (vehicle_age + np.random.normal(0, 0.001, 1000))*3000
vehicle_age = vehicle_age.reshape(-1, 1)    #single feature, use (1,-1) for single sample
vehicle_price = vehicle_price.reshape(-1, 1)

plt.scatter(vehicle_age, vehicle_price)
```

```python
### code to find out the relation. 


reg = LinearRegression().fit(vehicle_age, vehicle_price)
print(reg.score(vehicle_age, vehicle_price))  #returns the R-squared
print(reg.intercept_)
print(reg.coef_)
```

```python
fitLine = reg.predict(vehicle_age)
plt.scatter(vehicle_age, vehicle_price)
plt.plot(vehicle_age, fitLine, c='r')
plt.show()
```
